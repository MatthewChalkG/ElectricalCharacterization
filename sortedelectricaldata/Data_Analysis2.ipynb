```python
data = pd.read_csv("datalogging/Data/HallEffectDCBias/hallDCBiasSMU1681453699.txt")

def avgHallSignal(data, dcList):
    #dcList = np.linspace(-200, 200, 21)
    dataP = data[data['i'] == 3.2]
    dataN = data[data['i'] == -3.2]
    avgP = []
    avgN = []

    for dc in dcList:
        dataDC = dataP[abs(dataP['dc'] - dc) < .1]
        avgX = np.mean(dataDC['x'])
        avgP += [avgX]
                  
    for dc in dcList:
        dataDC = dataN[abs(dataN['dc'] - dc) < .1]
        avgX = np.mean(dataDC['x'])
        avgN += [avgX]
    
    avgDiff = [avgP[i]-avgN[i] for i in range(len(avgP))]
    return avgDiff
```
