```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def evenOdd(data):
    even = []
    odd = []
    for i in range(len(data['x'])):
        if i % 2 == 0:
            even += [list(data['x'])[i]]
        elif i % 2 == 1:
            odd += [list(data['x'])[i]]
    diff = [even[i] - odd[i] for i in range(len(odd))]
    return even, odd, diff
            
```

```python
def splitByCurrent(data):
    iData = list(data['i'])
    xData = list(data['x'])
    tData = list(data['t'])
    
    posCurrX = []
    negCurrX = []
    
    posCurrT = []
    negCurrT = []
    

    for j in range(len(iData)):
        if iData[j] > 0:
            posCurrX  += [xData[j]]
            posCurrT += [tData[j]]
        else:
            negCurrX += [xData[j]]
            negCurrT += [tData[j]]

    return posCurrX, negCurrX, posCurrT, negCurrT

def diffTwo(data):
    posCurrX, negCurrX, posCurrT, negCurrT = splitByCurrent(data)
    if len(posCurrX) > len(negCurrX):
        posCurrX = posCurrX[1:]
        posCurrT = posCurrT[1:]
    elif len(posCurrX) < len(negCurrX):
        negCurrX = negCurrX[1:]
        negCurrT = negCurrT[1:]
    
    diffX = [posCurrX[i] - negCurrX[i] for i in range(len(negCurrX))]
    offX = [(posCurrX[i] + negCurrX[i])/2 for i in range(len(negCurrX))]
    avgT = [(posCurrT[i] + negCurrT[i])/2 for i in range(len(negCurrX))]
    return avgT, diffX, offX
```

```python
data = pd.read_csv("datalogging/hallDCBias.txt")
a,b,c,d  = splitByCurrent(data)
plt.scatter(c, a, label = "+i")
plt.scatter(d, b, label = "-i")
plt.legend()
```
